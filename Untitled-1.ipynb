```python
import onnxruntime as ort

# Check available providers
print("Available providers:", ort.get_available_providers())

# Check if CUDA/GPU provider is available
if 'CUDAExecutionProvider' in ort.get_available_providers():
    print("CUDA provider is available")
else:
    print("CUDA provider not available - using CPU only")
```

```python
import onnxruntime as ort

# Check available execution providers
print("Available providers:", ort.get_available_providers())

# Try to create a session with GPU
try:
    # If you have multiple GPUs, you can specify which one to use with provider options:cite[9]
    options = ort.SessionOptions()
    session = ort.InferenceSession("your_model.onnx", sess_options=options, providers=['CUDAExecutionProvider'])
    print("Session providers:", session.get_providers())
    print("✅ Success! ONNX Runtime is using the GPU.")
except Exception as e:
    print("❌ Failed to create session with GPU:", e)
```
